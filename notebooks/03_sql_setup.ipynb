```python
import pandas as pd
import sqlite3
from pathlib import Path

df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors = "coerce")
df = df.dropna(subset=["TotalCharges"])

#Create a SQLite database file in the project root
db_path = Path("../telco_churn.db")
conn = sqlite3.connect(db_path)

#Write the table into SQLite
df.to_sql("telco_churn", conn, if_exists="replace", index=False)
# Quick sanity check: number of rows loaded
pd.read_sql_query("SELECT COUNT(*) AS n_rows FROM telco_churn;", conn)
```
